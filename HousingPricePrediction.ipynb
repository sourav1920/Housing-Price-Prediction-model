```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint
import pickle

# Load data
df = pd.read_csv('/content/data.csv')

# Display basic information
print(df.head())
print(df.describe())
print(df.isnull().sum())  # Checking for null values
print(df.duplicated().sum())  # Checking for duplicate values

# Check and handle outliers
sns.boxplot(x=df['price'])
plt.show()

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['price'] < (Q1 - 1.5 * IQR)) | (df['price'] > (Q3 + 1.5 * IQR)))]  # Handling outliers

sns.boxplot(x=df['price'])
plt.show()

sns.histplot(df['price'])
plt.show()

# Plotting features against price
fig = go.Figure()
for feature in df.columns[:-1]:  # Exclude the target variable 'price'
    fig.add_trace(go.Scatter(x=df[feature], y=df['price'], mode='lines+markers', name=feature))

fig.update_layout(
    title='Line Graph of Price vs Features',
    xaxis_title='Feature Value',
    yaxis_title='Price'
)
fig.show()

# Plot distributions of numerical features
numerical_features = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']

plt.figure(figsize=(20, 15))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# Correlation matrix
plt.figure(figsize=(12, 10))
corr_matrix = df[numerical_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Pairwise relationships
sns.pairplot(df[numerical_features].dropna())
plt.show()

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Plot price trend over time
plt.figure(figsize=(12, 6))
df.groupby(df['date'].dt.to_period('M'))['price'].mean().plot()
plt.title('Average Price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.show()

# Feature engineering
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df.drop(columns=['date'], inplace=True)

# Encode categorical features
categorical_features = df.select_dtypes(include=['object']).columns
label_encoders = {}
for column in categorical_features:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Prepare data for modeling
X = df.drop(columns=['price'])
y = df['price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.to_csv('X_scaled.csv', index=False)
y.to_csv('y.csv', index=False)
print(X_scaled_df.head())

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Linear Regression with Ridge Regularization
ridge_model = Ridge()
ridge_model.fit(X_train, y_train)
ridge_predictions = ridge_model.predict(X_test)
ridge_score = r2_score(y_test, ridge_predictions)
ridge_cv_score = cross_val_score(ridge_model, X_scaled, y, cv=5, scoring='r2').mean()

print(f'Ridge Regression R^2 Score: {ridge_score}')
print(f'Ridge Regression Cross-Validation R^2 Score: {ridge_cv_score}')

# Decision Tree Regressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_score = r2_score(y_test, dt_predictions)
dt_cv_score = cross_val_score(dt_model, X_scaled, y, cv=5, scoring='r2').mean()

print(f'Decision Tree R^2 Score: {dt_score}')
print(f'Decision Tree Cross-Validation R^2 Score: {dt_cv_score}')

# ANN (MLPRegressor) with Hyperparameter Tuning
ann_model = MLPRegressor(hidden_layer_sizes=(50,), max_iter=500, random_state=42)
ann_model.fit(X_train, y_train)
ann_predictions = ann_model.predict(X_test)
ann_score = r2_score(y_test, ann_predictions)
ann_cv_score = cross_val_score(ann_model, X_scaled, y, cv=5, scoring='r2').mean()

print(f'ANN R^2 Score: {ann_score}')
print(f'ANN Cross-Validation R^2 Score: {ann_cv_score}')

# Hyperparameter Tuning for ANN using Grid Search
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],
    'max_iter': [500, 1000],
    'activation': ['relu', 'tanh']
}
grid_search = GridSearchCV(MLPRegressor(random_state=42), param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)
best_ann_model = grid_search.best_estimator_
best_ann_predictions = best_ann_model.predict(X_test)
best_ann_score = r2_score(y_test, best_ann_predictions)
best_ann_cv_score = cross_val_score(best_ann_model, X_scaled, y, cv=5, scoring='r2').mean()

print(f'Optimized ANN R^2 Score: {best_ann_score}')
print(f'Optimized ANN Cross-Validation R^2 Score: {best_ann_cv_score}')
print(f'Best Hyperparameters for ANN: {grid_search.best_params_}')

# Randomized Search for Random Forest
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(1, 30),
    'min_samples_split': randint(2, 10)
}
random_search = RandomizedSearchCV(RandomForestClassifier(), param_distributions=param_dist, n_iter=50, cv=5, scoring='accuracy', random_state=42)
random_search.fit(X_train, np.round(y_train))  # Ensure target is integer for classification

print(f'Best Parameters (RandomizedSearchCV): {random_search.best_params_}')
print(f'Best Score (RandomizedSearchCV): {random_search.best_score_}')

# Save the trained model to a pickle file
with open('random_forest_model.pkl', 'wb') as file:
    pickle.dump(random_search, file)

print("Model saved to random_forest_model.pkl")

# Verify the loaded model by making predictions
predictions = random_search.predict(X_test)
print("Predictions:", predictions)
```
